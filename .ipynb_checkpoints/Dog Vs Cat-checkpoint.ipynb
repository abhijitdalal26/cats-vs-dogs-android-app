## 1. Predicting using trained model

```python
import tensorflow as tf
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
model_path = "./model/cat_vs_dog_model.keras"
model = tf.keras.models.load_model(model_path)
```

```python
test_dr = "./data/test/"
test_img = os.listdir(test_dr)[1] # 100.jpg
test_img_path = os.path.join(test_dr, test_img)
print(test_img_path)
```

```python
img = tf.keras.utils.load_img(test_img_path, target_size=(224, 224))
img_array = tf.keras.utils.img_to_array(img)
img_array = tf.expand_dims(img_array, 0)  # Model expects a batch (1, 224, 224, 3)
img_array = tf.keras.applications.mobilenet_v3.preprocess_input(img_array)
```

```python
logit = model.predict(img_array)
print(logit)
```

```python
probability = tf.nn.sigmoid(logit).numpy()[0][0]
print(probability)
```

```python
if probability > 0.5:
    print(f"Prediction: DOG ({probability:.2%})")
else:
    print(f"Prediction: CAT ({1 - probability:.2%})")
```

```python
plt.imshow(img)
plt.axis("off")
plt.show()
```

## 2. Converting to TensorFlow Lite

```python
model.summary()
```

```python
for i, layer in enumerate(model.layers):
    print(f"Layer {i}: {layer.name}")

# we dont need that Layer 1 i.e Data augmentaion layer 
```

```python
# the .get_layer(), uses the same layer objects (already trained ).
mobilenet = model.get_layer("MobileNetV3Large")
gap = model.get_layer("global_average_pooling2d_1")
dropout = model.get_layer("dropout_1")
classifier = model.get_layer("dense")
```

```python
IMG_SIZE = (224, 224)
inputs = tf.keras.Input(shape=IMG_SIZE + (3,))

x = mobilenet(inputs)
x = gap(x)
x = dropout(x, training=False)   # Dropout OFF for inference
outputs = classifier(x)

inference_model = tf.keras.Model(inputs, outputs) 
```

```python
saved_model = "./model/saved_model"
# model.export(saved_model)
```

```python
converter = tf.lite.TFLiteConverter.from_saved_model(saved_model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
```

```python
interpreter = tf.lite.Interpreter(
    model_path="/model/model",
    experimental_delegates=[]   # important for INT8 on desktop
)
interpreter.allocate_tensors()
```
